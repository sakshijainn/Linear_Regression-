```python
# Import the dependencies / libraries
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
#Create a dataframe
df = pd.read_csv('DecisionTree.csv')
```

```python
df
```

```python
X = df.iloc[:,:-1]
```

```python
X
```

```python
y = df.iloc[:,6]
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder_X = LabelEncoder()
```

```python
X = X.apply(LabelEncoder().fit_transform)
```

```python
X
```

```python
regressor = DecisionTreeClassifier()
```

```python
regressor.fit(X.iloc[:,1:6],y)
```

```python
X_in = np.array([1,1,0,0,1])
```

```python
y_pred = regressor.predict([X_in])
```

```python
y_pred
```
